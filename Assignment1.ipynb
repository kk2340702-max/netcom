```python
FILE = "students.txt"

f = open(FILE, "a")
f.close()


def grade(mark):
    if mark >= 90:
        return "A+"
    elif mark >= 80:
        return "A"
    elif mark >= 70:
        return "B"
    elif mark >= 60:
        return "C"
    else:
        return "D"


def add_student():
    f = open(FILE, "a")

    sid = input("Enter ID: ")
    name = input("Enter Name: ")
    course = input("Enter Course: ")
    email = input("Enter Email: ")
    phone = input("Enter Phone: ")
    marks = input("Enter Marks: ")

    data = sid + "," + name + "," + course + "," + email + "," + phone + "," + marks + "\n"

    f.write(data)
    f.close()

    print("Student Added Successfully")


def view_students():
    f = open(FILE, "r")

    print("\nID\tNAME\tCOURSE\tMARKS\tGRADE")

    for line in f:
        s = line.strip().split(",")

        print(s[0], "\t", s[1], "\t", s[2], "\t", s[5], "\t", grade(int(s[5])))

    f.close()


def search_student():
    sid = input("Enter Student ID: ")

    found = 0

    f = open(FILE, "r")

    for line in f:
        s = line.strip().split(",")

        if s[0] == sid:
            print("\nStudent Found")
            print("ID:", s[0])
            print("Name:", s[1])
            print("Course:", s[2])
            print("Email:", s[3])
            print("Phone:", s[4])
            print("Marks:", s[5])
            print("Grade:", grade(int(s[5])))
            found = 1

    f.close()

    if found == 0:
        print("Student Not Found")


def update_student():
    sid = input("Enter Student ID: ")

    f = open(FILE, "r")
    lines = f.readlines()
    f.close()

    f = open(FILE, "w")

    found = 0

    for line in lines:
        s = line.strip().split(",")

        if s[0] == sid:
            found = 1

            name = input("Enter New Name: ")
            course = input("Enter New Course: ")
            email = input("Enter New Email: ")
            phone = input("Enter New Phone: ")
            marks = input("Enter New Marks: ")

            newline = sid + "," + name + "," + course + "," + email + "," + phone + "," + marks + "\n"

            f.write(newline)

        else:
            f.write(line)

    f.close()

    if found == 1:
        print("Record Updated")
    else:
        print("Student Not Found")


def delete_student():
    sid = input("Enter Student ID: ")

    f = open(FILE, "r")
    lines = f.readlines()
    f.close()

    f = open(FILE, "w")

    found = 0

    for line in lines:
        s = line.strip().split(",")

        if s[0] == sid:
            found = 1
        else:
            f.write(line)

    f.close()

    if found == 1:
        print("Record Deleted")
    else:
        print("Student Not Found")


def statistics():
    f = open(FILE, "r")

    marks = []

    for line in f:
        s = line.strip().split(",")
        marks.append(int(s[5]))

    f.close()

    total = len(marks)

    if total > 0:
        print("Total Students:", total)
        print("Highest Mark:", max(marks))
        print("Lowest Mark:", min(marks))
        print("Average Mark:", sum(marks) / total)


def export_report():
    f = open(FILE, "r")

    marks = []
    topper = ""
    highest = 0

    for line in f:
        s = line.strip().split(",")

        mark = int(s[5])

        marks.append(mark)

        if mark > highest:
            highest = mark
            topper = s[1]

    f.close()

    total = len(marks)
    average = sum(marks) / total

    r = open("report.txt", "w")

    r.write("STUDENT REPORT\n")
    r.write("Total Students: " + str(total) + "\n")
    r.write("Average Marks: " + str(average) + "\n")
    r.write("Topper: " + topper + "\n")

    r.close()

    print("Report Exported")


while True:

    print("\nSTUDENT MANAGEMENT SYSTEM")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Statistics")
    print("7. Export Report")
    print("8. Exit")

    ch = int(input("Enter Choice: "))

    if ch == 1:
        add_student()

    elif ch == 2:
        view_students()

    elif ch == 3:
        search_student()

    elif ch == 4:
        update_student()

    elif ch == 5:
        delete_student()

    elif ch == 6:
        statistics()

    elif ch == 7:
        export_report()

    elif ch == 8:
        print("Exiting........")
        break

    else:
        print("Invalid Choice")
```
